
## 7-2 檔案存取



Python 使用 open()函式以及檔案物件的 close()方法來開啟與關閉檔案。

檔案物件所提供的 read()、readline()或 readlines()方法可用來讀取檔案裡的資料; 

write()方法則是用來將資料寫入檔案中。



### 7-2-1 讀取檔案的內容



在下面的例子中，我們利用 open()開啟一個已經存存的 news.txt 檔案，
再透過檔案物件所提供的 read()來讀取檔案內容.

其中檔案開啟的模式如下:

* 'r' - 讀取(檔案需存在)
* 'w' - 新建檔案寫入(檔案可不存在，若存在則清空)
* 'a' – 在舊檔案裡新增新的文字內容(若檔案不存在則新建該檔案)


```python
import os

f = open("07-news.txt", 'r')
f.read()
```


針對上面的寫法，若使用 with 敘述也可以得到相同的功能:

(注意兩者的不同點)


```python

with open("07-news.txt", 'r') as f: 
    print(f.read())
    
f.close()
    
```


### 7-2-2 將資料寫入檔案




在下面的例子中，我們利用 open()開啟一個原本不存在的 07-newfile1.txt 檔案，

再透過檔案物件所提供的 write()將指定的字串寫入檔案裡:


把檔案打開看看:



```python

with open("07-newfile1.txt", 'r') as f: 
    print(f.read())
    
f.close()
```


如果我們想在已建立的檔案裡，從文字最後方加上新的內容，則可以在 open() 函式裡設定使用'a'模式來開啟檔案，此處我們以剛建立的 newfile1.txt 當作例子:


```python
f = open("07-newfile1.txt", 'a') 
f.write('Python程式設計')
f.close()
```


我們在資料夾裡打開 newfile1.txt，查看裡面的內容，會發現增加了一句新的文字內容:



```python

with open("07-newfile1.txt", 'r') as f: 
    print(f.read())
    
f.close()
```


如果不想新增的字串直接與原有資料連在一起，我們可以利用 \n 來換行，

例如:


```python
f = open("07-newpoem.txt", 'a')

f.write('白日依山盡\n')
f.write('黃河入海流\n')
f.write('欲窮千里目\n')
f.write('更上一層樓\n')

f.close()
```


由於資料夾裡並沒有 newpoem.txt 這個檔案，所以第一行程式會建立這個檔案，並且依序在新增指定的字串後換行，再增加下一個字串。

執行結果如下:


```python

with open("07-newpoem.txt", 'r') as f: 
    print(f.read())
    

f.close()
```


### 7-2-3 逐行讀出檔案內容



對於上面所建立之 newpoem.txt，我們可以使用 readline()方法從檔案讀取一 行資料，

例如:



```python
f = open("07-newpoem.txt", 'r') 

line1 = f.readline()
line2 = f.readline()
line3 = f.readline()
line4 = f.readline()

print(line1) 
print(line2) 
print(line3) 
print(line4)

f.close()
```


請注意上面的換行間隔





Python 另外提供了 readlines()來讀取檔案裡的所有行，並且利用串列的形式來回傳所有的行內容，每一行字串都會成為串列中的一個元素，例如:


```python
f = open("07-newpoem.txt", 'r') 

line = f.readlines()
print(line)

f.close()
```


### 7-2-4 其他逐行讀出檔案內容的方法




對於之前所建立之 newpoem.txt，我們可以利用 for-loop 可以逐行讀取檔案的內容，例如:



```python

f = open("07-newpoem.txt", 'r')

for line in f: 
    
    print(line)
    
f.close()
```


另外，我們也可以使用流程控制搭配 readline()的方式來逐行讀取讀取檔案中 的文字資料，例如:



```python


f = open("07-newpoem.txt", 'r') 

line = f.readline()

while line: 
    print(line)
    line = f.readline()
        
        
    
```


注意一下, 為何在迴圈中要先 print(line) 再 line = f.readline()




當然，使用 for-loop 也可以達到相同的目的，例如:
    
    

```python

for line in open("07-newpoem.txt", 'r'): 
    print(line)
    
```


而除了利用 readline()來讀取一行的資料以外，我們也可以用 readlines()產生所有文件中所有行的串列內容，再利用 for-loop 逐一列印出來，例如:


```python

f = open("07-newpoem.txt", 'r')

lines = f.readlines()

for line in lines: 
    print(line)
    
```

csv(comma-separated values，逗點分隔值)是一種簡單的文字檔格式，它以逗號(也可使用其他特殊符號)來分隔不同欄位的資料;第一行通常是用來描述 資料內容的表頭(header)，不過這是可有可無的。

若要在 Python 中讀取或產生 csv 的檔案，可以使用內建的 csv 標準模組。




### 7-3-1 讀取 CSV 檔案的內容



我們從網路上下載一個 07-108_student.csv，其每一行的資料之間是以逗號分隔的，檔案的內容如下:



```python

with open("07-108_student.csv", 'r') as f: 
    print(f.read())
    
f.close()
    
```


透過下面的程式內容便可讀取這個 csv 檔案的內容:



```python
import csv
f = open("07-108_student.csv", newline='')

lines = csv.reader(f)

for l in lines: 
    print(l)
```


此程式會開啟 csv 檔，之後透過 reader()函式來解析檔案的內容，把每一行的資料用逗號切開，再回傳一個串列。




若使用 with 敘述，則程式寫法如下:


```python
import csv

with open("07-108_student.csv", newline='') as f:
    lines = csv.reader(f)
    for l in lines: 
        print(l)
        
```

自然語言處理的其中一個重要環節就是斷詞的處理。

英文「Power outage all over Taiwan」，可以直接用空白斷成「Power / outage / all / over / Taiwan」;
但中文斷詞在先天上就比較難處理，以停電新聞來說，「全台大停電」在斷字處理上 就可以做出「全台 / 大停電」或是「全 / 台大 / 停電」等不同的語意，而不同的斷句就會有不同結果。




Jieba 結巴是基於 python 的開源中文斷詞套件，可以用於中文的斷詞應用。

site: https://pypi.org/project/jieba/

加入台灣常用語詞的版本:

https://github.com/APCLab/jieba-tw

安裝方式:

pip install git+https://github.com/APCLab/jieba-tw.git


### 7-3-1 斷詞



使用 jieba 斷詞功能時，得先引入這個套件:


```python
import jieba

seg_list = jieba.cut("台灣每過了一分鐘，在非洲就過了六十秒")
print("|".join(seg_list))
```


結巴的斷詞分成兩個模式，精確模式(default mode)只找到系統演算法認為最 精準的斷詞可能;

而全斷詞模式(full mode)則是把所有可以成詞的詞語都列出來, 速度較快，但不能解決歧義。

如果我們想要用來做文本分析，而句子又不是很長 的的時候，使用全斷詞模式也許可以增加文跟文章之間的可比性:

Ex:


```python

seg_list = jieba.cut("我來到桃園銘傳大學", cut_all=False) 
    
print("Default Mode: " + "/ ".join(seg_list)) #精確模式

seg_list = jieba.cut("我來到桃園銘傳大學", cut_all=True) 

print("Full Mode: " + "/ ".join(seg_list)) # 全模式
```


### 7-3-2 自定字典 userdict.txt



Jieba也允許開發者指定自己自定義的字典，以便包含 jieba 詞庫里沒有的語詞;

雖然 jieba 有新詞識別能力，但是自行添加新語詞可以保證更高的正確率。

此處，我們先進行一個簡單的斷詞:


```python

seg_list = jieba.cut("我參加銘傳大學的研究計畫")

print("/ ".join(seg_list))
```


如果我們覺得斷詞的效果不好，例如希望將

"我參加銘傳大學的研究計畫"

斷詞為

"我 / 參加 / 銘傳大學 / 的 / 研究計畫/"，

那麼可以利用自定字典 userdict.txt 來達到目的。

首先，我們建立一個 userdict.txt 的檔案，並將之儲存(在 記事本中儲存時選擇編碼: UTF-8) 在目前的工作目錄中，內容如下:




userdict.txt 的格式必須一個詞佔一行;每一行分三部分:詞語、詞頻(可省略)、詞性(可省略)，用空格隔開，而且順序不可顛倒。接下來在程式碼裡新 增一行載入自訂字典的程式碼:



```python
jieba.load_userdict("userdict.txt")
seg_list = jieba.cut("我參加銘傳大學的研究計畫")
print(" / ".join(seg_list))
```


我們可以發現斷詞的結果從

"我/ 參加/ 銘傳/ 大學/ 的/ 研究/ 計畫"

改變為

"我 / 參加 / 銘傳大學 / 的 / 研究計畫"




### 7-3-3 讀取文件與分析文件中重要的字詞




TF-IDF 是一種用於資訊檢索與文本探勘(text mining)的技術，它能評估一個語詞對於一份文章的重要程度。

當我們手頭有一些文章時，我們可以利用 TF-IDF 技術來自動地進行關鍵詞擷取的統計任務。

jieba 使用了 tf-idf 方法來實作萃取 出文章中關鍵詞的功能:



```python

#載入jieba與jieba.analyse。

import jieba
import jieba.analyse

# 把檔案的內容讀出來，請將文章存在目錄的07-news中

f = open('07-news.txt','r',encoding = 'utf8') 

article = f.read()
# 透過jieba內建的idf頻率庫, 我們可以計算文章中最重要的字詞 

tags = jieba.analyse.extract_tags(article, 100) 

print("最重要字詞", tags)
```


程式中的 jieba.analyse.extract_tags(article, 100)，就是告訴 jieba 我們要從這 個文章中取出前 100 個 tf-idf 值最大的關鍵詞。
